```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
import seaborn as sns
```

# Population   

```python

# Load your data into a DataFrame
df = pd.read_csv('data/pop_dk.csv', sep=';', thousands=',', index_col='Age group')

# Create separate dataframes for males and females
males = df.iloc[:, -2]
females = df.iloc[:, -1]

# Make the male distribution negative for the pyramid effect
males = males.mul(-1)

# Plot the data
fig, ax = plt.subplots()
ax.barh(df.index, males, color='red', label = "males")
ax.barh(df.index, females, color='gray', label = "females")

# Format x-axis numbers to include commas and remove negative sign
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(abs(x))))

# Set x-axis limits
max_pop = max([max(females), max(males)])
ax.set_xlim(-max_pop*1.1, max_pop*1.1)

ax.set_xlabel('Population')
ax.set_ylabel('Age-Group')
ax.set_title('Denmark Population Pyramid')

# Add a legend
ax.legend()
plt.show()
```

```python
import pandas as pd

# Load the data
data = pd.read_csv('data/pop.csv', delimiter=';')

# Define a function to convert age groups to mid-points
def age_to_midpoint(age_group):
    if age_group == '110 years +':
        return 110
    else:
        age_group = age_group.replace(' years', '')
        lower, upper = age_group.split('-')
        return (int(lower) + int(upper)) / 2
    # Add more conditions here for other age groups

# Apply the function to the 'Age' column
data['Age'] = data['Age'].apply(age_to_midpoint)

# Calculate the total population for each row
data['Total'] = data['Men'] + data['Women']

# Calculate the weighted age for each row
data['Weighted_Age'] = data['Age'] * data['Total']

# Calculate the mean age for each area
mean_age = data.groupby('Area')['Weighted_Age'].sum() / data.groupby('Area')['Total'].sum()


# Filter mean age for multiple cities
cities = ['All Denmark', 'Copenhagen', 'Lyngby-Taarbk', 'Aarhus', 'Aalborg']  # Replace with your city names
mean_age_cities = mean_age.loc[cities]

mean_age_cities
```
